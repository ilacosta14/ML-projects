We want to separate the object data in ordered and unordered data

```python
ode_cols = ['LotShape', 'LandContour','Utilities','LandSlope',  'BsmtQual',  'BsmtFinType1', 'BsmtFinType2' , 'CentralAir',  'Functional', 'GarageFinish', 'GarageQual', 'PavedDrive', 'ExterCond', 'KitchenQual', 'BsmtExposure', 'HeatingQC','ExterQual', 'BsmtCond', 'GarageCond','Fence'] #data we will handle with ordinal encoding
ohe_cols = ['Street','Alley','LotConfig','Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd','Foundation',  'Electrical',  'SaleType', 'SaleCondition', 'Heating', 'GarageType', 'RoofMatl', 'MSZoning'] #data we will handle with one-hot encoding
```

#### Build Pipeline to transofrm categorical data

```python
from sklearn.compose import make_column_transformer, ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

ode_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
col_trans = ColumnTransformer(transformers=[
    ('ode_p', ode_pipeline, ode_cols),
    ('ohe_p', ohe_pipeline, ohe_cols),
    ],
    remainder='passthrough', 
    n_jobs=-1)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessing', col_trans)
])
```

## Define variables x and y

```python
X = train_dataset.drop('SalePrice',axis=1) #define x, putting all the rows and all the columns except the last one
y = train_dataset['SalePrice'] #define y, putting all the rows but only the last column

X_p = pipeline.fit_transform(X) #X preprocessed
print(X_p)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_p,y, test_size = 0.2, random_state = 42)
```

## Training the Random Forest Regression model on the whole dataset

```python
param_grid = { 'max_depth': [5, 10, 15], 'n_estimators': [100, 250, 500], 'min_samples_split': [3, 5, 10] }
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(random_state = 42) # Create a Random Forest Regressor
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
regressor_cv = GridSearchCV(regressor, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
regressor_cv.fit(X_train, y_train)
np.sqrt(-1 * regressor_cv.best_score_) 
```

## Apply model on the test dataset

```python
test_dataset
```

```python
data_test_preprocess = pipeline.transform(test_dataset)
estimator = regressor_cv.best_estimator_
y_stacking = np.exp(estimator.predict(data_test_preprocess))

df_y_stacking_out = test_dataset[['Id']]
df_y_stacking_out['Id'] = df_y_stacking_out['Id'].astype(int)
df_y_stacking_out['SalePrice'] = y_stacking

df_y_stacking_out.to_csv('submission.csv', index=False)
```
